### Learning Objectives

```
#@title ðŸ”§ Setup
# We'll use 'experta' (a Python rule engine) to model forward chaining.

import sys, subprocess                                      # For optional installs
def pip_install(pkgs):
    for p in pkgs:
        try: __import__(p.split("==")[0])                   # Try importing
        except Exception:
            subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", p])  # Install quietly
pip_install(["experta"])                                    # Install the rule engine

from experta import KnowledgeEngine, Rule, Fact, Field, P    # Import key classes
print("âœ… Setup complete for Week 9.")
```

```
#@title ðŸ§  Course advising expert system (fully commented)

class Student(Fact):                                         # A fact describing a student
    gpa = Field(float, mandatory=False)                      # Optional GPA slot
    likes_math = Field(bool, mandatory=False)                # Likes mathematics?
    likes_programming = Field(bool, mandatory=False)         # Likes programming?

class Adviser(KnowledgeEngine):                              # Our rule engine
    @Rule(Student(gpa=P(lambda g: g is not None and g >= 4.0), likes_programming=True))
    def recommend_ai(self):                                  # Rule 1: high GPA + likes programming â†’ AI/ML
        print("Recommend: Advanced AI/ML electives.")

    @Rule(Student(likes_math=True, likes_programming=False))
    def recommend_stats(self):                               # Rule 2: math but not programming â†’ Statistics/DS
        print("Recommend: Statistics/Data Science pathway.")

    @Rule(Student())
    def default(self):                                       # Rule 3: general default advice
        print("Recommend: Meet an adviser for a tailored plan.")

engine = Adviser()                                           # Create the engine
engine.reset()                                               # Reset working memory
engine.declare(Student(gpa=4.2, likes_math=True, likes_programming=True))  # Insert a fact
engine.run()                                                 # Fire rules
```
